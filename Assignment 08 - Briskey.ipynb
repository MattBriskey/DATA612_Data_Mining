```python
import pandas as pd

# the mpg data is used for demonstration
df = pd.read_csv("https://raw.githubusercontent.com/franklin-univ-data-science/data/master/mpg.csv")
df.head()
```

```python
def summarize_data(g):
  return(
    pd.Series({
      "avg_MPG": g.mpg.mean(),
      "min_MPG": g.mpg.min(),
      "max_MPG": g.mpg.max(),
      "sample_count": len(g),
      "mix": g.cylinders.sum()/df.cylinders.sum(),
    })
  )

df.groupby("cylinders").apply(summarize_data)
```

### Read the summarize_data() and explain the difference between g and df.

df is the dataframe while g is the aggregated data grouped by number of cylinders.  df has 406 different rows of data and g groups all of the data by number of cylinders to get mean, min, max, count, and % of total for each grouped number of cylinders.  For example, there are 207 rows of data in df that have 4 cylinders.  g groups all of these 207 rows into an aggregrate group and then calculates the mean, min, max, count, and % of total for that specific group.

```python
# Custom function to return the avg, min, max, count, and % of total grouped by Artist
def summarize_data(g):
  return(
    pd.Series({
      "avg_streams": g.Streams_Billions.mean(),
      "min_streams": g.Streams_Billions.min(),
      "max_streams": g.Streams_Billions.max(),
      "sample_count": len(g),
      "mix": g.Artist.count()/df2.Artist.count(),
    })
  )

df3 = df2.groupby('Artist').apply(summarize_data)
df5 = df3.sort_values(by='sample_count', ascending=False) #Sort the resulting dataframe in descending order by mix

format_dict = {
    'avg_streams':'{0:,.2f}', 
    'min_streams': '{0:,.2f}',  
    'max_streams': '{0:,.2f}',
    'sample_count': '{0:,.0f}',
    'mix': '{:.2%}'}
df5.style.format(format_dict).highlight_max(color = 'lightgreen', axis = 0) # highlights the max value of each column in green
```
