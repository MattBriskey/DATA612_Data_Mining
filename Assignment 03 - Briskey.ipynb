```python
# load the data
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/stock_data/stocks_yahoo.csv', low_memory=False)
df.head(5)
```

```python
# convert the 'date' column to datetime format

df['date']= pd.to_datetime(df['date'])
```

```python
# Check to make sure the date conversion worked as intended
df.head(5)
```

```python
# Find the maximum date of your data.

max_date = df['date'].max()
max_date
```

```python
# add maximum date to data frame as a column

df['max_date'] = pd.Series([max_date for x in range(len(df.index))])
df.head(5)
```

```python
# Subtract all the dates on the column from the maximum date

date_diff_days = df['date'] - max_date
date_diff_days.head(5)
```

```python
# add date diff in days as a column to the data frame

df['date_diff_days'] = df['date'] - max_date
df.head(5)
```

```python
# Convert the number of days into number of months and save it as a new column

df['date_diff_months'] = df['date_diff_days'].\
    astype('timedelta64[M]')

df.head(5)
```

```python
# Save the dataset as a csv file

df.to_csv('../COMP 612 Computing for Data Analytics/Assignment03_df.csv')
```
