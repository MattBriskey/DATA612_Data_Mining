```python
# load the data
import pandas as pd
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/chendaniely/pandas_for_everyone/master/data/housing.csv', low_memory=False)
df.head(5) #Shows the first 5 rows
```

### Histogram

Histograms are used to gorup data into bins and then summarize the data by count.  The shape of the histogram displays the distribution of the data

```python
# Plot histogram

sns.set() #Changes the background color to dark and adds grid lines
sns.histplot(data = df
             , x = 'Expense.per.SqFt'
             , kde = True # kde is the kernel density estimate which is the trended line along the histogram
             , bins = 50 # changes the number of bins to 50
             , alpha = 0.5 # sets the transparency of each column.  Acceptable values: 0-1, 0 is more transparent
             , hue = 'Boro') # sets each Boro to be a different color

```

### Hexbin Joint Plot

Hexbin plots are useful to represent the relationship of 2 numerical variables when you have a lot of data points.  They use color gradients to show density of the data and the color of each hexbin denotes the number of points in it.  A joint plot adds combines multiple plots into one.  In this case, the bhistogram is added to the X and Y axises.

```python
# Hexbin plot

hexbin = sns.jointplot(x="Expense.per.SqFt" #Determines the x-axis values
                       , y="Market.Value.per.SqFt" #Determines the y-axis vvalues
                       , data=df #Determines what data to use
                       , kind="hex" #A hexplot
                       , marginal_kws=dict(bins=30, fill=True)) #30 bins for the histogram and fills them with color

hexbin.set_axis_labels(xlabel='Expense Per Square Foot', ylabel='Market Value Per Square Foot') #Sets the axis labels

hexbin.fig.suptitle('Hexbin Joint Plot of Expense and Market Value Per Square Foot', #

                 fontsize=10, y= 1.03)
```

### Box Plot

Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups. They are built to provide high-level information at a glance, offering general information about a group of data's symmetry, skew, variance, and outliers.

```python
# Box plot

sns.boxplot(x="Boro" #Determines the x-axis values
            , y="Market.Value.per.SqFt" #Determines the y-axis values
            , data=df) #Determines what data to use
            
```

### Violin Plot

Violin plots are used to show distribution of numeric data, and are especially useful when you want to make a comparison of distributions between multiple groups. The peaks, valleys, and tails of each group's density curve can be compared to see where groups are similar or different.

```python
# Violin plot

df_boro_filtered= df.query("Boro in ('Bronx','Brooklyn','Staten Island')") # Filter rows based on specific Boro's

sns.violinplot(x="Boro" #Determine the x-values
               , y="Market.Value.per.SqFt" #Determine the y-values
               , data=df_boro_filtered #Determines the data\
               ,palette=['b','r','m']) #Sets the colors for each item
```
