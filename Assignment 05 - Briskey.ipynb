## Merging Data

```python
# load the data
import pandas as pd
import numpy as np

#Load Movie Info into dataframe 1
df1 = pd.read_csv('https://raw.githubusercontent.com/MattBriskey/data612_data_mining/main/data_sets/Movie_Info.csv', low_memory=False)
df1.head(10) #Shows the first 10 rows
```

```python
# Load Movie Ratings data into dataframe 2
df2 = pd.read_csv('https://raw.githubusercontent.com/MattBriskey/data612_data_mining/main/data_sets/Movie_Ratings.csv', low_memory=False)
df2.head(10) #Shows the first 10 rows
```

```python
df3 = pd.merge(df1, df2, on='id', how='left')  # Join the two dataframes together
df3.head(10) #Shows the first 10 rows
```

```python
# Count of the number of missing values
print(np.count_nonzero(df3['imdb_rating'].isnull()))
```

## Cleaning Missing Data

###  Recode/Replace

```python
# Replace NA values with 0
df4 = df3.fillna(0).iloc[0:250, 0:7]
df4.head(10) #Shows the first 10 rows
```

### Fill Forward

```python
# Fill Forward.  Missing values are replaced with the last known/recorded value
df5 = df3.fillna(method='ffill').iloc[0:250, 0:7]
df5.head(10) #Shows the first 10 rows
```

### Fill Backward

```python
# Fill Bakcward.  Missing values are replaced with the newest value
df6 = df3.fillna(method='bfill').iloc[0:250, 0:7]
df6.head(10) #Shows the first 10 rows
```
